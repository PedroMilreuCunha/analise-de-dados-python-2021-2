## Exemplo 03

- Vamos fazer um programa que informe ao aluno sua condição de aprovação mediante a nota final informada.

````
nota = float(input('Informe sua média final: '))

# Escopo global
print(nota)

# Segmentar escopos - regra de decisão
# Primeiro escopo
if nota >= 7:
  print('Parabéns! Você está aprovado(a).')

# Outro escopo
if nota < 7:
  print('Lamento infomar que você está reprovado(a).')
````

## else 

Quando uma decisão lógica é simplesmente o resultado complementar de condições anteriores podemos usar a cláusula `else`.

```
if <condição 1>:
    escopo de programação se a condição 1 for verdadeira
else:
    escopo de programação se a condição 1 for falsa
```

- Portanto, os blocos `if-else` permitem delimitar escopos de programação mediante validade de condições, que chamamos de regra (árvore) de decisão.



````
nota = float(input('Informe sua média final: '))

# Segmentar escopos - regra de decisão
if nota >= 7:
  # Escopo local
  print('Parabéns! Você está aprovado(a).')
else:
  # Escopo local
  print('Lamento infomar que você está reprovado(a).')
````

````
# Vamos solicitar as seguintes informações e criar variáveis
renda = float(input('Informe sua renda mensal: '))
bolsista = bool(int(input('Digite 1 se for bolsista e 0 se c.c.: ')))
aposentado = bool(int(input('Digite 1 se for aposentado e 0 se c.c.: ')))

# Usar a estrutura de decisão if-else não agrupada
# Testar a seguinte condição lógica com operadores de relacionais
if renda > 1200 and bolsista is False and aposentado is False:
  print('Você deve pagar imposto!')
else:
  print('Você possui isenção!')
````

````
# Vamos solicitar as seguintes informações e criar variáveis
renda = float(input('Informe sua renda mensal: '))

# Usar a estrutura de decisão if-else agrupada ou aninhada
# Testar a seguinte condição lógica com operadores de relacionais
if renda > 1200:
  # Primeiro escopo
  bolsista = bool(int(input('Digite 1 se for bolsista e 0 se c.c.: ')))
  aposentado = bool(int(input('Digite 1 se for aposentado e 0 se c.c.: ')))
  # Segundo escopo
  if bolsista is False and aposentado is False:
     # Escopo se verdadeira
    print('Você deverá pagar imposto!')
  else:
    print('Você possui isenção!')
else:
  print('Você possui isenção!')
````

## Exemplo 06

Escreva um programa que calcule o preço a pagar pelo fornecimento de energia elétrica. Pergunte a quantidade de energia elétrica consumida (kWh) e o tipo de instalação: *R* para residências, *C* para comércios e *I* para indústrias. Considere os dados abaixo com o preço por tipo e faixa de consumo:

|Tipo|Faixa (kWh) | Preço|
|:---:|:---:|:---:|
Residencial| Até 500 | 0,40|
Residencial| Acima de 500 | 0,65|
Comercial| Até 1000 | 0,55|
Comercial| Acima 1000 | 0,60|
Industrial| Até 5000 | 0,55|
Industrial| Acima 5000 | 0,60|

````
# Variáveis com valores iniciais
consumo = float(input("Informe a quantidade consumida de energia elétrica kwh: "))
tipo = input("Digite R se for um cliente residencial, C se for comercial e I se for industrial: ")


# Escopo e testes lógicos aninhados
if tipo == 'R':
  # Escopo do tipo == 'R'
  cliente = 'A sua residência'
  if consumo <= 500:
    # Escopo local
    tarifa = 0.40
  else:
    # Escopo local
    tarifa = 0.65

# Outro escopo if
if tipo == 'C':
  cliente = 'O seu comércio'
  if consumo <= 1000:
    tarifa = 0.55
  else:
    tarifa = 0.60

# Outro escopo if
if tipo == 'I':
  cliente = 'A sua indústria'
  if consumo <= 5000:
    tarifa = 0.55
  else:
    tarifa = 0.60


# Escopo global
# Informando o resultado
# Resultados
valor = consumo * tarifa

print(f"""{cliente} teve um consumo de {consumo} kWh de energia neste mês. 
A tarifa para essa faixa de consumo é de R$ {tarifa:.2f}.
Dessa forma, a sua conta foi de R$ {valor:.2f} """) 
````

# elif (else if)

Podemos usar uma estrutra de decisão do tipo `if ... else if ... else`. Geralmente, esse estrutura funcionada bem quando precisamos fazer **múltiplos testes envolvendo um mesmo objeto**.

O `Pyhton` apresenta uma solução interessante para o caso de múltiplas condições "se" `if`, evitando a criação de outro nível de estrutura condicional. Assim, usamos a condição "se não ... se" `elif`.

Captura de Tela 2020-09-03 às 9.11.51 AM.png


```
if <condição 1>:
    escopo caso a condição 1 seja verdadeira.
elif <condição 2>:
    escopo caso a condição 2 seja verdadeira dado que a condição 1 é falsa
else:
    escopo caso as condições 1 e 2 sejam falsas 
```

## Exemplo 07

Vamos revisitar o programa de energia elétrica usando o encademento de condições `if-elif-else`.

````
# Variáveis com valores iniciais
consumo = float(input("Informe a quantidade consumida de energia elétrica kwh: "))
tipo = input("Digite R se for um cliente residencial, C se for comercial e I se for industrial: ")

# Escopo e testes lógicos aninhados
if tipo == 'R':
  # Escopo do tipo == 'R'
  cliente = 'A sua residência'
  if consumo <= 500:
    # Escopo local
    tarifa = 0.40
  else:
    # Escopo local
    tarifa = 0.65

# Outro escopo elif
elif tipo == 'C':
  cliente = 'O seu comércio'
  if consumo <= 1000:
    tarifa = 0.55
  else:
    tarifa = 0.60

# Outro escopo else
else:
  cliente = 'A sua indústria'
  if consumo <= 5000:
    tarifa = 0.55
  else:
    tarifa = 0.60


# Escopo global
# Informando o resultado
# Resultados
valor = consumo * tarifa

print(f"""{cliente} teve um consumo de {consumo} kWh de energia neste mês. 
A tarifa para essa faixa de consumo é de R$ {tarifa:.2f}.
Dessa forma, a sua conta foi de R$ {valor:.2f} """) 
````

## Exemplo 08


Faça um programa para calcular o IMC de uma pessoa e informar seu atual status nutricional.

O IMC é calculado por peso/(altura*altura) e pode trazer os seguintes resultados:

| Intervalo | Condição| 
|:------:| :--------:|
Menos do que 18,5 |	Abaixo do peso |
Entre 18,5 e 24,9 |	Peso normal |
Entre 25 e 29,9	| Sobrepeso |
Entre 30 e 34,9	| Obesidade grau 1 |
Entre 35 e 39,9	| Obesidade grau 2 |
Mais do que 40	| Obesidade grau 3 |

````
# Solicitar os dados do paciente
nome = str(input("Informe o seu nome:"))

# Solicitar os dados do paciente
altura = float(input("Informe sua altura:"))

# Solicitar os dados do paciente
peso = float(input("Informe seu peso:"))

# Calcular o IMC 
imc = peso/(altura*altura)

# Criar variável condição com estrutura/regras de decisão
# Já adiantamos alguns operadores lógicos ( >= , <=, > , and, etc)
if imc < 18.5:
  # Escopo local
  condicao = "abaixo do peso"
elif imc <= 24.9:
  condicao = "peso normal"
elif imc <= 29.9:
  condicao = "sobrepeso"
elif imc <= 34.9:
  condicao = "obesidade grau 1"    
elif imc <= 39.9:
  condicao = "obesidade grau 2"  
else:
  condicao = "obesidade grau 3" 

# Escopo global
# Reportar resultado
print(f"""{nome} registra peso de {peso:.1f} kg e altura de {altura:.1f} m. 
      O seu IMC é {imc:2.f}. Portanto, sua condição é {condicao} """ )
````

## Exemplo 11

Faça um programa para um caixa eletrônico. O programa deverá perguntar ao usuário o valor do saque e depois informar quantas notas de cada valor serão
fornecidas. As notas disponíveis são 1, 5, 10, 50 e 100 reais. O valor mínimo para saque é de 10 reais e o valor máximo de 600 reais. O programa não deve se preocupar com a quantidade de notas exisitentes na máquina.

````
# Perguntar o valor do saque.
saque = float(input("Digite o valor do saque: "))

# Testar a condição principal
if 10 <= saque <= 600:

  # Escopo local
  # Divisão com resultado inteiro //
  notas_cem = saque // 100
  saque = saque % 100
  notas_cinquenta = saque // 50
  saque = saque % 50
  notas_dez = saque // 10
  saque = saque % 10
  notas_cinco = saque // 5
  saque = saque % 5
  notas_um = saque // 1

  if notas_cem > 0:
    print(f"{notas_cem:.0f} notas de R$ 100.")
  if notas_cinquenta > 0:
    print(f"{notas_cinquenta:.0f} notas de R$ 50.")
  if notas_dez > 0:
    print(f"{notas_dez:.0f} notas de R$ 10.")
  if notas_cinco > 0:
    print(f"{notas_cinco:.0f} notas de R$ 5.")
  if notas_um > 0:
    print(f"{notas_um:.0f} notas de R$ 1.")  

else:
  print("Não é possível realizar o saque.")  

````
